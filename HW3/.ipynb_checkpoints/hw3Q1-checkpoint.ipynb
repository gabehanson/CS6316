```python
import pandas as pd
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import numpy as np
from numpy import random 
import matplotlib.pyplot as plt
import sklearn
from tqdm import tqdm
import time
import seaborn as sns
import math
  
%matplotlib inline
```

```python
r = np.linspace(-10, 10, 100)
z = 1/(1 + np.exp(-r))
  
plt.plot(r, z)
plt.title('Sigmoid (r)')
plt.xlabel("r")
plt.ylabel("Sigmoid(r)")
  
plt.show()
```

```python
r = np.linspace(-10, 10, 100)
dzdr = (1/(1 + np.exp(-r)))*(1 - (1/(1 + np.exp(-r))))
  
plt.plot(r, dzdr)
plt.title('Derivative of Sigmoid (r)')
plt.xlabel("r")
plt.ylabel("d Sigmoid(r)/dr")
  
plt.show()
```
