```python
# load the data
import mnist_loader
(training_data, validation_data, test_data) = mnist_loader.load_data_wrapper()
training_data = list(training_data)
validation_data = list(validation_data)
test_data = list(test_data)
```

```python
# prepping the data for use
X_train = [i for i, j in training_data]
X_train = np.concatenate(X_train).ravel()
X_train = X_train.reshape(50000,784).astype("float32")

y_train = [j for i, j in training_data]
y_train = np.concatenate(y_train).ravel()
y_train = y_train.reshape(50000,10).astype("float32")

X_test = [i for i, j in test_data]
X_test = np.concatenate(X_test).ravel()
X_test = X_test.reshape(10000,784).astype("float32")

y_test = [j for i, j in test_data]
#y_test = np.concatenate(y_test).ravel()
#y_test = y_test.reshape(10000,1).astype("float32")
```

## simple model 2 hidden units per layer

```python
from keras.models import Sequential  
from keras.layers import *  

model = keras.Sequential()

model.add(Input(shape=(784,))) # Input tensor
model.add(Dense(units=2, activation = 'sigmoid', name="hidden_layer_1"))# hidden layer 1
model.add(Dense(units=2, activation="sigmoid", name="hidden_layer_2"))#hidden layer 2
model.add(Dense(units=10, activation="softmax", name = 'predictions'))#output layer 

model.summary()
```

```python
loss_fn = keras.losses.SparseCategoricalCrossentropy()
model.compile(loss= tf.losses.categorical_crossentropy, optimizer='adam', metrics=['accuracy',
    tf.keras.metrics.MeanSquaredError(),
    tf.keras.metrics.AUC()])                                            
```

```python
print("Fit model on training data")
history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    batch_size=10,
    epochs=2,
)
```

```python
y_test = pd.Series(y_test)

def one_hot_encode(series):
    return pd.get_dummies(series).values


y_test = one_hot_encode(y_test)
```

```python
results1 = model.evaluate(X_test, y_test)
```

### adjust learning rates

```python
from matplotlib.pyplot import figure


def fit_model(trainX, trainy, testX, testy, lrate):
    # define model
    model = keras.Sequential()
    model.add(Input(shape=(784,))) # Input tenso
    model.add(Dense(units=300, activation = 'sigmoid', name="hidden_layer_1"))# hidden layer 
    model.add(Dense(units=150, activation="sigmoid", name="hidden_layer_2"))#hidden layer 
    model.add(Dense(units=10, activation="softmax", name = 'predictions'))#output layer
    
    # compile model
    loss_fn = keras.losses.SparseCategoricalCrossentropy()
    opt =  tf.keras.optimizers.Adam(learning_rate = lrate)
    model.compile(loss= tf.losses.categorical_crossentropy, optimizer=opt, metrics=['accuracy',
    tf.keras.metrics.MeanSquaredError(),
    tf.keras.metrics.AUC()])         
    # fit model
    history = model.fit(trainX, trainy, validation_data=(testX, testy), epochs=20, batch_size = 100, verbose=0)
    # plot learning curves
    plt.plot(history.history['accuracy'], label='train')
    plt.plot(history.history['val_accuracy'], label='test')
    plt.legend()
    plt.title('lrate='+str(lrate)+'   Final test accuracy ='+str(history.history['val_accuracy'][-1]), pad=-50)
    plt.xticks(ticks = [*range(20)], labels = [*range(1,21,1)])
    plt.ylim([0, 1])
    plt.xlabel('# of epochs')
    print('final accuracy: ', history.history['accuracy'])
```

```python
# create learning curves for different learning rates
learning_rates = [1E-1, 1E-2, 5E-3, 1E-3, 5E-4, 1E-4, 1E-5, 1E-6, 5E-7]

t = time.time()

time 

plt.figure(figsize=(20, 14)) 
for i in range(len(learning_rates)):
    # determine the plot number
    plot_no = (330 + (i+1))
    plt.subplot(plot_no)
    print('training model for learning rate: ', learning_rates[i])
    # fit model and plot learning curves for a learning rate
    fit_model(X_train, y_train, X_test, y_test, learning_rates[i])
# show learning curves
plt.tight_layout()
plt.savefig('rates.png')
plt.show()

elapsed = time.time() - t

print('This run took :', elapsed, ' seconds')
```

```python
# the step size doesn't seem to have much of an effect on the accuracy, rather, it has an effect on how many epochs it takes to converge
```
